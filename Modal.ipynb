```python
import time
t0 = time.time()
import json
from pathlib import Path
import panel as pn

## Load config:
config_path = Path('config.json')
if not config_path.is_file():
    !python update_config.py
if config_path.is_file():
    with open(config_path,encoding='utf-8') as config_file:
        config = json.load(config_file)
else:
    raise ValueError("config.json doesn't exist!")

## Set CSS:
pn.extension(
    raw_css=[config["css"].replace('\\n','').replace('\n','')],
    loading_spinner='dots',
    loading_color='#00aa41'
)

## First load: just the dashboard, help buttons, and language.
dash = pn.template.VanillaTemplate(title="Climate Analogues | Analogues climatiques")


w_sidetitle = pn.pane.Markdown('##Start new search')
dash.sidebar.append(w_sidetitle)
w_loading = pn.indicators.LoadingSpinner(value=True,width=100,height=100)
dash.sidebar.append(pn.Row(pn.layout.HSpacer(),w_loading,pn.layout.HSpacer()))

## MODAL: 
w_about_en = pn.pane.Markdown('# About')
w_about_fr = pn.pane.Markdown('#  Propos')
modal_lang = pn.Row(pn.layout.HSpacer(),w_about_en,pn.layout.HSpacer(),w_about_fr,pn.layout.HSpacer(),min_width=600)
dash.modal.append(modal_lang)

def open_modal(event):
    dash.open_modal()
    
## HEADER:
w_open_modal = pn.widgets.Button(name='About |  Propos', width = 150)
w_open_modal.on_click(open_modal)
pn.state.onload(dash.open_modal)

LOCALE = "en"
w_language = pn.widgets.Button(name="Franais", width=150)
w_headerbox = pn.Row(pn.layout.HSpacer(),w_open_modal, w_language)
dash.header.append(w_headerbox)

def change_language(event):
    global LOCALE
    LOCALE = "fr" if (LOCALE == "en") else "en"
    w_language.name = "English" if LOCALE == "fr" else "Franais"
    dash.title = {"en":"Climate Analogues of Canadian Cities", "fr":"Analogues climatiques de villes canadiennes"}[LOCALE]
    w_sidetitle.object = {"en":"##Start new search","fr":"##Dbutez nouvelle recherche"}[LOCALE]
w_language.on_click(change_language)
dash.servable()

t1 = time.time()
print('Time to first load:',t1 - t0)

w_loading.value = False
```
